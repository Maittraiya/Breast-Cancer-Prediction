```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, ExtraTreesClassifier

def models(X_train, Y_train):
    # Logistic Regression
    log = LogisticRegression(random_state=0)
    log.fit(X_train, Y_train)

    # Decision Tree
    tree = DecisionTreeClassifier(random_state=0, criterion="entropy")
    tree.fit(X_train, Y_train)

    # Support Vector Machine (SVM)
    svm = SVC(random_state=0, kernel="linear")
    svm.fit(X_train, Y_train)

    # Random Forest
    rf = RandomForestClassifier(random_state=0, n_estimators=10)
    rf.fit(X_train, Y_train)

    # Gradient Boosting
    gb = GradientBoostingClassifier(random_state=0)
    gb.fit(X_train, Y_train)

    # AdaBoost
    ada = AdaBoostClassifier(random_state=0)
    ada.fit(X_train, Y_train)

    # Extra Trees
    et = ExtraTreesClassifier(random_state=0)
    et.fit(X_train, Y_train)

    return log, tree, svm, rf, gb, ada, et
```

```python
from sklearn.metrics import accuracy_score

# Get the models
log, tree, svm, rf, gb, ada, et = models(X_train, Y_train)

# Predictions for each model
log_pred = log.predict(X_test)
tree_pred = tree.predict(X_test)
svm_pred = svm.predict(X_test)
rf_pred = rf.predict(X_test)
gb_pred = gb.predict(X_test)
ada_pred = ada.predict(X_test)
et_pred = et.predict(X_test)

# Accuracy for each model
log_acc = accuracy_score(Y_test, log_pred)
tree_acc = accuracy_score(Y_test, tree_pred)
svm_acc = accuracy_score(Y_test, svm_pred)
rf_acc = accuracy_score(Y_test, rf_pred)
gb_acc = accuracy_score(Y_test, gb_pred)
ada_acc = accuracy_score(Y_test, ada_pred)
et_acc = accuracy_score(Y_test, et_pred)

# Print the accuracy for each model
print("Logistic Regression Accuracy: {:.2f}%".format(log_acc * 100))
print("Decision Tree Accuracy: {:.2f}%".format(tree_acc * 100))
print("SVM Accuracy: {:.2f}%".format(svm_acc * 100))
print("Random Forest Accuracy: {:.2f}%".format(rf_acc * 100))
print("Gradient Boosting Accuracy: {:.2f}%".format(gb_acc * 100))
print("AdaBoost Accuracy: {:.2f}%".format(ada_acc * 100))
print("Extra Trees Accuracy: {:.2f}%".format(et_acc * 100))
```
